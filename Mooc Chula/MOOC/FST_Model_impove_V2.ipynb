```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets, linear_model
```

```python
file_path = r"C:\Users\Pannawit\Documents\GitHub\Deep-learning\NeuronNetwork\NN1\housing.csv"

# Define the column headers
headers = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]

# Load the data with proper delimiter handling
housing_data = pd.read_csv(file_path, delim_whitespace=True, names=headers)
print(housing_data)
```

```python
y = housing_data['MEDV']
X = pd.DataFrame(housing_data['RM'])
```

```python
# Visualize data
plt.scatter(X, y)
plt.xlabel("RM")
plt.ylabel("price")
```

```python
print(housing_data.head())

print(housing_data.describe())

print(housing_data.isnull().sum())

print(housing_data.corr())

import seaborn as sns

sns.pairplot(housing_data)
plt.show()

plt.figure(figsize=(10, 8))
sns.heatmap(housing_data.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
model = linear_model.LinearRegression()

model.fit(X,y)
```

```python
print(f"Price = {float(model.coef_[0])}*RM + {float(model.intercept_)}")
```

```python
Y_hat = model.coef_[0]*X + model.intercept_
```

```python
lineX = np.arange(4, 9, 0.001)
lineY = model.coef_[0]*lineX + model.intercept_
```

```python
plt.scatter(X, y)
plt.scatter(lineX, lineY, color='red', s=1)
#plt.scatter(X, Y_hat, color='red', s=2)

# Add labels and title for better clarity
plt.xlabel('RM')
plt.ylabel('Predicted Price')
plt.title('Scatter Plot of Predicted Prices vs RM')
```
