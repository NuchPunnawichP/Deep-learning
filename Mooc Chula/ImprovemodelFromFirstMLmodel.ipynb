```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

from sklearn import datasets, linear_model
```

```python
from sklearn.datasets import fetch_california_housing

california = fetch_california_housing()

X, y = california.data, california.target
```

```python
california_df = pd.DataFrame(X, columns=california.feature_names)
california_df['MedHouseVal'] = y

plt.figure(figsize=(10, 6))
sns.scatterplot(data=california_df, x='MedInc', y='MedHouseVal', alpha=0.7)
plt.title('Median Income (MedInc) vs. Median House Value (MedHouseVal)')
plt.xlabel('Median Income (in $10,000s)')
plt.ylabel('Median House Value (in $100,000s)')
plt.grid(True)
plt.show()
```

```python
model = linear_model.LinearRegression()

model.fit(X, y)
```

```python
print("y^ = ", model.coef_[0], "X + ", model.intercept_)
```

```python
y_pred = model.coef_ * X + model.intercept_
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y_pred)
plt.title('Median Income (MedInc) vs. Median House Value (MedHouseVal)')
plt.xlabel('Median Income (in $10,000s)')
plt.ylabel('Median House Value (in $100,000s)')
plt.grid(True)

plt.show()
```
