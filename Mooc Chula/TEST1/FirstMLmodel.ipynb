```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets, linear_model
```

```python
dataset = datasets.load_breast_cancer()
```

```python
dataset.keys()
```

```python
print(dataset['DESCR'])
```

```python
y = dataset['target']

x = dataset['data'][:, [5]]
```

```python
plt.scatter(x, y)
```

```python
model = linear_model.LinearRegression()

model.fit(x, y)
```

```python
print("y^ = ", model.coef_[0], "X + ", model.intercept_)
```

```python
y_pred = model.coef_ * x + model.intercept_

plt.scatter(x, y)
plt.scatter(x, y_pred, color = 'red', linewidths=0.5)
```
