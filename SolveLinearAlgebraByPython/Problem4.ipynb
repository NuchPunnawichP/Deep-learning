```python
import numpy as np
import matplotlib.pyplot as plt

def plot_lines(data):
    for line_data in data:
        x = line_data['x']
        y = line_data['y']
        
        # Ensure x and y are numpy arrays
        x = np.array(x)
        y = np.array(y)
        
        plt.plot(x, y, label=line_data.get('label', 'Line'))
    plt.legend()
    plt.show()
```

```python
A = np.array([ [6, 1], [5, 2] ])
b = np.array([18, 22])

X = np.hstack((A, b.reshape((2, 1))))
print(X)
```

```python
try:
    x = np.linalg.solve(A, b)
    print('Non-singular Matrix')
    print(f'The intersection point is {tuple(x.astype(int))}')
except np.linalg.LinAlgError as err:
    print(err)
```
